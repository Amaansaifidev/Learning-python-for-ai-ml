```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
x = ["Monday", "Tuesday", "Wednesday","Thursday","Friday", "Saturday", "Sunday"]
```

```python
y = [10, 20, 12, 18, 16, 15, 18]
```

```python
plt.plot(x, y)
plt.title("bakery shop sales")
plt.xlabel("days", fontsize=20, color="red")
plt.ylabel("sales", fontsize=20, color="red")
plt.show()
```

```python
plt.plot(x, y, color="green")
plt.title("bakery shop sales")
plt.xlabel("days", fontsize=20, color="red")
plt.ylabel("sales", fontsize=20, color="red")
plt.show()
```

```python
plt.plot(x, y, color="green", linestyle="--", label="Dashed")
plt.title("bakery shop sales")
plt.xlabel("days", fontsize=20, color="red")
plt.ylabel("sales", fontsize=20, color="red")
plt.show()
```

```python
plt.plot(x, y, color="green", linestyle="-", label="Solid")
plt.title("bakery shop sales")
plt.xlabel("days", fontsize=20, color="red")
plt.ylabel("sales", fontsize=20, color="red")
plt.show()
```

```python
dataframe = {
"Year":[2010, 2011, 2012, 2013],
"Sales":[200, 230, 210, 200]}
```

```python
df = pd.DataFrame(dataframe)
```

```python
plt.plot(df["Year"], df["Sales"], color="darkblue", linestyle="--",linewidth="3", marker="o", label="Dashed")
plt.xlabel("Sales")
plt.ylabel("years")
plt.grid()
plt.legend()
plt.show()
```

```python
labels = ['East', 'West', 'South', 'North']
sizes = [40, 30, 10, 20]  # Percent values

plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
plt.title('Programming Language Popularity', fontsize=14, color='darkblue')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle
plt.show()
```
